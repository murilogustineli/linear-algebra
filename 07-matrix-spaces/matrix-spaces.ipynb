## Column space of a matrix
- **Notation:** $C(A)$
- **Definition:** the vector subspace spanned by the column of A

$\large
\underset{\text{A set of vectors}}{\left\{\begin{bmatrix}
1\\
2\\
7
\end{bmatrix}\begin{bmatrix}
2\\
4\\
8
\end{bmatrix}\right\}} \Rightarrow \underset{A\ matrix}{\begin{bmatrix}
1 & 2\\
2 & 4\\
7 & 8
\end{bmatrix}}
$

**Formal definition:** Column space is the set of all vectors that can be obtained by any weighted linear combination of all the columns in the matrix.

$\large
 \begin{array}{l}
C( A) =\{\beta _{1} a_{1} +...+\beta _{n} a_{n}\} ;\ \ \ \beta \in \mathbb{R}\\
C( A) =span(\{a_{1} ,...,a_{n}\})
\end{array}
$

Example:

$\large
 \begin{array}{l}
\begin{bmatrix}
-3\\
1\\
5
\end{bmatrix} \in C\left(\begin{bmatrix}
3 & 0\\
5 & 2\\
1 & 2
\end{bmatrix}\right)\\
\\
\left[ -1\begin{bmatrix}
3\\
5\\
1
\end{bmatrix} +3\begin{bmatrix}
0\\
2\\
2
\end{bmatrix}\right] =\begin{bmatrix}
-3\\
1\\
5
\end{bmatrix}
\end{array}
$

**Important question in linear algebra: $v\in C(A)$?**

$\large
v\in C( A)\text{?} \Rightarrow \begin{cases}
\text{YES} \Rightarrow \text{What coefficients?}\\
\text{NO} \Rightarrow \text{How close is it to} \ C( A)\text{?}
\end{cases}
$

Example:

$\large
\begin{bmatrix}
-3\\
1\\
5
\end{bmatrix} =-1\begin{bmatrix}
3\\
5\\
1
\end{bmatrix} +3\begin{bmatrix}
0\\
2\\
2
\end{bmatrix} \Rightarrow \underset{\text{3x2}}{\begin{bmatrix}
3 & 0\\
5 & 2\\
1 & 2
\end{bmatrix}}\underset{\text{2x1}}{\begin{bmatrix}
-1\\
3
\end{bmatrix}} =\underset{\text{3x1}}{\begin{bmatrix}
-3\\
1\\
5
\end{bmatrix}}
$

## Column space, visualized in code

```python
# Matrix S
S = np.array([[3,0], [5,2], [1,2]])

# Vector v
v = np.array([[-3], [1], [5]])
# v = np.array([[1], [7], [3]])

fig = plt.figure(figsize=(5,5), dpi=100)
ax = fig.add_subplot(projection='3d')

# Draw plane corresponding to the column space
xx, yy = np.meshgrid(np.linspace(-5,5,10), np.linspace(-5,5,10))
cp = np.cross(S[:,0], S[:,1])
z1 = (-cp[0]*xx - cp[1]*yy)/cp[2]
ax.plot_surface(xx, yy, z1, alpha=.2)


# Plot the two vectors from matrix S
ax.plot([0, S[0,0]], [0, S[1,0]], [0, S[2,0]], 'k')
ax.plot([0, S[0,1]], [0, S[1,1]], [0, S[2,1]], 'k')

# And the vector v
ax.plot([0, v[0]], [0, v[1]], [0, v[2]], 'r')

ax.view_init(elev=20, azim=0)
plt.show();
```

## Row space of a matrix
The **row space** of a matrix is the subspace spanned by the rows of the matrix.

Notation:

$\large
 \begin{array}{l}
R( A) =C\left( A^{T}\right)\\
\\
\overset{\text{Colum space}}{\begin{bmatrix}
3 & 0\\
5 & 2\\
1 & 2
\end{bmatrix}\begin{bmatrix}
-1\\
3
\end{bmatrix} =\begin{bmatrix}
-3\\
1\\
5
\end{bmatrix}} \ \ \ \ \ \ \ \ \ \ \overset{\text{Row space}}{\begin{bmatrix}
-1 & 0 & 3
\end{bmatrix}\begin{bmatrix}
3 & 0\\
5 & 2\\
1 & 2
\end{bmatrix} =\begin{bmatrix}
0 & 6
\end{bmatrix}}
\end{array}
$

## Null space and left null space of a matrix

Null space interpretation: it's the subspace such that any vector gets sent into the void of no return.

$N(A)$: The set of all vectors $v$ such that $Av = 0$ and $v\neq 0$

## Left null space of a matrix

$N(A^T)$: The set of all vectors $v$ such that $v^T A = 0^T$ and $v^T\neq 0^T$

Also written: $A^T v = 0$

## Column/left-null and row/null spaces are orthogonal
Recap.

Geometric definition of the dot product:

$\large
\alpha =|a||b|cos( \theta _{ab})
$

#### Column-space
vector is orthogonal to the column space of matrix A:

$\large
 \begin{array}{l}
v\perp C( A) \Rightarrow v^{T} a_{1} =0\\
\\
v^{T}\{\alpha _{1} a_{1} +...+\alpha _{n} a_{n}\} =0
\end{array}
$

Left-null space:

$
A^{T} v=0
$

The left-null space is orthogonal to the column space.

#### Row-space
vector is orthogonal to the row space of matrix A:

$\large
 \begin{array}{l}
v\perp R( A) \Rightarrow v^{T} a_{1}^{T} =0\\
\\
v^{T}\left\{\alpha _{1} a_{1}^{T} +...+\alpha _{n} a_{n}^{T}\right\} =0
\end{array}
$

## Dimensions of column/row/null spaces
Dimension $\neq$ rank

Example:

$\large
 \begin{array}{l}
A=\begin{bmatrix}
1 & 0 & 3\\
0 & 4 & -5\\
3 & 4 & 0
\end{bmatrix}\\
\\
rank( A) =3\\
N( A) =\{\}\\
dim( N( A)) =0\\
dim( C( A)) =3
\end{array}
$

## More on $Ax=b$ and $Ax=0$
$\large
Ax = b
$

$\large
AX=B\ \begin{cases}
\begin{matrix}
Ax_{1} =b_{1}\\
Ax_{2} =b_{2}\\
...\\
Ax_{n} =b_{n}
\end{matrix}
\end{cases}
$

We know the elements of $A$ and $b$. The goal is to find the vector $x$ that makes the equation true.

### $Ax = b$ in statistics

$\large
 \begin{array}{l}
X\beta =y\\
X=\text{Design matrix}\\
\beta =\text{Regression coefficients}\\
y=\text{Observed data}
\end{array}
$

<br>

### $Ax = b$: Is there an exact solution?

$\large
Ax = b
$

#### Solution? Yes, if $b \in C(A)$

$\large
x_{1}\begin{bmatrix}
a_{1,1}\\
a_{2,1}
\end{bmatrix} +x_{2}\begin{bmatrix}
a_{1,2}\\
a_{2,2}
\end{bmatrix} =\begin{bmatrix}
b_{1}\\
b_{2}
\end{bmatrix}
$

<br>

### $Ax = \hat{b}$: Is there an exact solution?

$\large
Ax = \hat{b}
$

#### Solution? Yes, if $\hat{b} \in C(A)$

### More on $Ax = 0$

$\large
 \begin{array}{l}
Ax=0\\
( A-\lambda I) x=0\\
\text{where:}\\
\lambda =\text{Eigenvalue}\\
x=\text{Eigenvector}
\end{array}
$

### 3. Write a program that can read in a list of words and convert each word into an integer, and a separate program that convert integers into words. 
- The program should work with words up to 10 characters in length.
- It should generate integers in the range 0 to N^10, where N is the number of characters in the alphabet you are using.
- If N=26, then your maximum value would be 26^10 = 141167095653376.

```python
def word_to_integer(word_list):
    """
    Program that reads a list of words and converts each word into an integer
    """
    int_list = []
    for word in word_list:
        num = 0
        if len(word) <= 10:
            for char in word.lower():
                num += (ord(char) - 96)**10
            int_list.append(num)
    return int_list
```

```python
def integer_to_word_func(num, prev=None):
    """
    Program that reads an integer and coverst into a word
    """
    # Digits
    mapper = {
        1: 'One',
        2: 'Two',
        3: 'Three',
        4: 'Four',
        5: 'Five',
        6: 'Six',
        7: 'Seven',
        8: 'Eight',
        9: 'Nine',
        10: 'Ten',
        11: 'Eleven',
        12: 'Twelve',
        13: 'Thirteen',
        14: 'Fourteen',
        15: 'Fifteen',
        16: 'Sixteen',
        17: 'Seventeen',
        18: 'Eighteen',
        19: 'Nineteen',
        20: 'Twenty',
        30: 'Thirty',
        40: 'Forty',
        50: 'Fifty',
        60: 'Sixty',
        70: 'Seventy',
        80: 'Eighty',
        90: 'Ninety',
        100: 'Hundred',
        1000: 'Thousand',
        1000000: 'Million',
        1000000000: 'Billion',
        1000000000000: 'Trillion'
    }
    
    # make a list of the map and descending order
    orderedmap = [(val, name) for val, name in sorted(mapper.items(), reverse=True)]
    
    if num == 0:
        return 'Zero'
    
    if num <= 20:
        if prev is None or prev >= 100:
            return [mapper[num]]
        else:
            return []
    
    # Store the result as a list (better perfomance than appending to a string)
    result = []
    for val, name in orderedmap:
        print(val, name)
        if val > num:
            continue
        ndivs, nmod = divmod(num, val)
        print(ndivs, nmod)
        # print(ndivs, nmod)
        result += integer_to_word_func(ndivs, val) + [name]
        print(result)
        num = nmod
        print('num:', num)
        
    return result

def int_to_word(num):
    return " ".join(integer_to_word_func(num))
```
